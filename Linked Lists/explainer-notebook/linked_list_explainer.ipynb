# Singly Linked List

A singly linked list is a linear data structure where elements are not stored at contiguous memory locations. Instead, each element (node) points to the next, forming a chain.

**Key Advantages:**
- Dynamic size (can grow/shrink easily).
- Efficient insertions and deletions at the head (O(1)).

## 1. Node Class Definition

Every linked list consists of nodes. Each node holds the `value` and a `next` pointer referencing the subsequent node in the list.

```python
class Node:
    """Represents a single node in a singly linked list.

    Attributes:
        value: The data stored in the node.
        next: A reference to the next node in the sequence (default is None).
    """
    def __init__(self, value):
        self.value = value
        self.next = None
```

## 2. Initialization and Traversal

The `LinkedList` class manages the nodes. It tracks the `head` (first node), `tail` (last node), and `length` of the list for O(1) appends and length lookups. We also include a `print_list` method to help us visualize the list during traversal.

```python
class LinkedList:
    """A singly linked list implementation.

    Attributes:
        head: A reference to the first node in the list.
        tail: A reference to the last node in the list.
        length: The current number of nodes in the list.
    """
    def __init__(self, value):
        """Initializes a new linked list with a single node.

        Args:
            value: The data for the initial node.
        """
        new_node = Node(value)
        self.head = new_node
        self.tail = new_node
        self.length = 1

    def print_list(self):
        """Prints all values in the linked list sequentially."""
        iterator = self.head
        # Traverse the list until reaching the end (None)
        while iterator:
            print(iterator.value)
            iterator = iterator.next
```

## 3. Insertion Operations (Append, Prepend, Insert)

Inserting a new node requires managing pointers. We can append to the end, prepend to the beginning, or insert at a specific index.

> *Note: For Jupyter Notebook demonstration purposes, we are assigning these functions to the `LinkedList` class dynamically.*

```python
def append(self, value):
    """Adds a new node to the end of the linked list.

    Args:
        value: The data for the new node.
    Returns:
        True to indicate successful append.
    """
    new_node = Node(value)
    # Case 1: Empty list
    if self.length == 0:
        self.head = new_node
        self.tail = new_node
    # Case 2: List has elements
    else:
        self.tail.next = new_node
        self.tail = new_node
    self.length += 1
    return True

def prepend(self, value):
    """Adds a new node to the beginning of the linked list.

    Args:
        value: The data for the new node.
    Returns:
        True to indicate successful prepend.
    """
    new_node = Node(value)
    
    # Case 1 - Empty list
    if not self.head:
        self.head = new_node
        self.tail = new_node
    # Case 2 - List has elements
    else:
        new_node.next = self.head
        self.head = new_node
        
    self.length += 1
    return True

def insert(self, index, value):
    """Inserts a new node at the specified index.

    Args:
        index: The position to insert the new node (0-indexed).
        value: The data for the new node.
    Returns:
        True if inserted successfully, False if index is out of bounds.
    """
    # Validate index bounds
    if index < 0 or index > self.length:
        return False
    
    if index == 0:
        return self.prepend(value)
    if index == self.length:
        return self.append(value)
    
    # Get the node immediately preceding the insertion point
    temp = self.get(index - 1)
    new_node = Node(value)
    
    # Wire the new node into the list
    new_node.next = temp.next
    temp.next = new_node
    self.length += 1
    return True

# Bind methods to the class
LinkedList.append = append
LinkedList.prepend = prepend
LinkedList.insert = insert
```

## 4. Search and Retrieval (Get & Set)

Finding a node by index requires traversing the list from the head.

```python
def get(self, index):
    """Retrieves the node at the specified index.

    Args:
        index: The position of the node to retrieve.
    Returns:
        The Node object at the index, or None if invalid.
    """
    # Edge cases - Out of bounds, negative index, or empty list
    if (self.head != None) and (index >= self.length):
        return None
    elif index < 0 or self.head == None:
        return None
    else:
        iterator = self.head
        iterator_index = 0
        # Traverse until the desired index is reached
        while iterator_index != index:
            iterator = iterator.next
            iterator_index += 1
        return iterator

def set_value(self, index, value):
    """Updates the value of the node at the specified index.

    Args:
        index: The position of the node to update.
        value: The new value to assign.
    Returns:
        True if update was successful, False otherwise.
    """
    temp = self.get(index)
    if temp:
        temp.value = value
        return True
    return False

# Bind methods to the class
LinkedList.get = get
LinkedList.set_value = set_value
```

## 5. Deletion Operations (Pop, Pop First, Remove)

Removing nodes requires careful pointer management to prevent memory leaks and broken chains.

```python
def pop(self):
    """Removes and returns the last node's value from the list.

    Returns:
        The value of the removed node, or None if empty.
    """
    # Case 1 - Empty list
    if not self.head:
        return None
    
    # Case 2 - One element
    elif self.head == self.tail:
        temp = self.head
        self.head = None
        self.tail = None
        self.length -= 1
        return temp.value
    
    # Case 3 - Multiple elements
    else:
        pointer = self.head
        # Traverse to the second-to-last node
        while pointer.next is not self.tail:
            pointer = pointer.next

        temp = self.tail
        self.tail = pointer
        self.tail.next = None
        self.length -= 1
        return temp.value

def pop_first(self):
    """Removes and returns the first node from the list.

    Returns:
        The removed Node object, or None if empty.
    """
    # Case 1 - Empty list
    if self.length == 0:
        return None
    
    # Case 2 - One element
    elif self.length == 1:
        temp = self.head
        self.head = None
        self.tail = None
        self.length -= 1
        return temp
    
    # Case 3 - Multiple elements
    else:
        temp = self.head
        # Shift head pointer to the next node
        self.head = self.head.next
        self.length -= 1
        return temp

def remove(self, index):
    """Removes the node at the specified index.

    Args:
        index: The position of the node to remove.
    Returns:
        The removed Node object, or None if invalid index.
    """
    if index < 0 or index >= self.length:
        return None
    if index == 0:
        return self.pop_first()
    if index == self.length - 1:
        return self.pop()
    
    # Get the node preceding the target node
    temp = self.get(index - 1)
    next_node = temp.next
    
    # Bypass the target node
    temp.next = next_node.next
    next_node.next = None
    self.length -= 1
    return next_node

# Bind methods to the class
LinkedList.pop = pop
LinkedList.pop_first = pop_first
LinkedList.remove = remove
```

## 6. Advanced Utility Operations

Common interview algorithms involving linked lists, such as reversing the list in-place, finding the middle node using slow/fast pointers, detecting cycles (loops), and removing duplicates.

```python
def reverse(self):
    """Reverses the linked list in-place in O(n) time and O(1) space."""
    # Swap head and tail
    temp = self.head
    self.head = self.tail
    self.tail = temp

    after = temp.next
    before = None

    # Iteratively flip the pointers
    for _ in range(self.length):
        after = temp.next
        temp.next = before
        before = temp
        temp = after

def find_middle_node(self):
    """Finds the middle node using a slow and fast pointer approach.

    Returns:
        The value of the middle node, or None if empty.
    """
    if self.length == 0:
        return None
        
    slow = self.head
    fast = self.head
    # Fast pointer moves twice as fast; when it reaches the end, slow is at the middle
    while fast is not None and fast.next is not None:
        fast = fast.next.next
        slow = slow.next
    return slow.value

def has_loop(self):
    """Detects if the linked list contains a cycle (Floyd's Cycle-Finding Algorithm).

    Returns:
        True if a cycle exists, False otherwise.
    """
    slow = self.head
    fast = self.head
    while fast is not None and fast.next is not None:
        fast = fast.next.next
        slow = slow.next
        if slow is fast:
            return True
    return False
    
def find_kth_from_end(self, k):
    """Finds the k-th node from the end of the list without using the length property.

    Args:
        k: The position from the end.
    Returns:
        The k-th Node from the end, or None if k is out of bounds.
    """
    slow = self.head
    fast = self.head
    
    # Create a gap of k between the fast and slow pointer
    for _ in range(k):
        if fast is not None:
            fast = fast.next
        else:
            return None # k is larger than the list size
            
    # Move both pointers until fast reaches the end
    while fast is not None:
        fast = fast.next
        slow = slow.next
    return slow

def remove_duplicates(self):
    """Removes duplicate values from the linked list.
    Retains the first occurrence of each value.

    Returns:
        The modified LinkedList instance.
    """
    if self.length <= 1:
        return self
        
    list_of_keys = {}
    temp = self.head
    after_temp = self.head.next

    # Edge case: List of size 2 with duplicates
    if (after_temp.value == temp.value) and self.length == 2:
        temp.next = None
        self.tail = temp
        self.length = 1
        return self

    list_of_keys[temp.value] = True

    while after_temp is not None:
        if after_temp.value not in list_of_keys:
            list_of_keys[after_temp.value] = True
            temp = temp.next
            after_temp = after_temp.next
        else:
            # Duplicate found, bypass the node
            duplicate_node = after_temp
            temp.next = temp.next.next
            after_temp = after_temp.next
            duplicate_node.next = None
            self.length -= 1

            if temp.next is None:
                self.tail = temp
    return self

# Bind methods to the class
LinkedList.reverse = reverse
LinkedList.find_middle_node = find_middle_node
LinkedList.has_loop = has_loop
LinkedList.find_kth_from_end = find_kth_from_end
LinkedList.remove_duplicates = remove_duplicates
```

## Examples and Testing

Now we test the functionality of our implementation.

```python
my_ll = LinkedList(1)
my_ll.append(2)
my_ll.append(3)
my_ll.append(6)

print("Current List:")
my_ll.print_list()

print("\nDoes the list have a loop?:", my_ll.has_loop())

print("\nTest: Remove Duplicates")
ll_1 = LinkedList(1)
for val in [2, 1, 3, 2, 4]:
    ll_1.append(val)
print("Before removing duplicates:", end=" ")
ll_1.print_list()

ll_1.remove_duplicates()
print("After removing duplicates: ", end=" ")
ll_1.print_list()
```

## Big O Time & Space Complexity Summary

| Operation | Best Case Time | Average Case Time | Worst Case Time | Space Complexity |
| :--- | :---: | :---: | :---: | :---: |
| **Append** | $\mathcal{O}(1)$ | $\mathcal{O}(1)$ | $\mathcal{O}(1)$ | $\mathcal{O}(1)$ |
| **Prepend** | $\mathcal{O}(1)$ | $\mathcal{O}(1)$ | $\mathcal{O}(1)$ | $\mathcal{O}(1)$ |
| **Insert** | $\mathcal{O}(1)$ | $\mathcal{O}(n)$ | $\mathcal{O}(n)$ | $\mathcal{O}(1)$ |
| **Pop** | $\mathcal{O}(n)$ | $\mathcal{O}(n)$ | $\mathcal{O}(n)$ | $\mathcal{O}(1)$ |
| **Pop First** | $\mathcal{O}(1)$ | $\mathcal{O}(1)$ | $\mathcal{O}(1)$ | $\mathcal{O}(1)$ |
| **Remove** | $\mathcal{O}(1)$ | $\mathcal{O}(n)$ | $\mathcal{O}(n)$ | $\mathcal{O}(1)$ |
| **Get / Lookup**| $\mathcal{O}(1)$ | $\mathcal{O}(n)$ | $\mathcal{O}(n)$ | $\mathcal{O}(1)$ |
| **Reverse** | $\mathcal{O}(n)$ | $\mathcal{O}(n)$ | $\mathcal{O}(n)$ | $\mathcal{O}(1)$ |
| **Remove Duplicates**| $\mathcal{O}(n)$ | $\mathcal{O}(n)$ | $\mathcal{O}(n)$ | $\mathcal{O}(n)$ |

*Note:* `pop()` operates in $\mathcal{O}(n)$ time because finding the node immediately preceding the tail requires traversing the list from the head (due to the lack of a `prev` pointer in a singly linked list).
